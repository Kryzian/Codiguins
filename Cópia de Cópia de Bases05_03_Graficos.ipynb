## A Base de Dados

````
import pandas as pd 
#https://drive.google.com/file/d/1_c1rtxf2YDaBKeQA62Gg4yjljuVO6mgc/view?usp=sharing
codigo = "1_c1rtxf2YDaBKeQA62Gg4yjljuVO6mgc"
file = "https://drive.google.com/u/3/uc?id=" + codigo + "&export=download"
milsa = pd.read_csv(file, sep = ";",decimal=",", index_col = "Funcionario")
print(milsa)
````

**Observação:**  Note o parâmetro index_col='Funcionario' utilizado na abertura do arquivo. Escolhemos a coluna "Funcionario" como a primeira a ser exibida. Assim, a numeração das linhas desapareceu.

A seguir a classificação de todas as variáveis desse conjunto de dados. 

* Funcionario	- Quantitativa discreta
* EstCivil - Qualitativa nominal
* Instrucao -	Qualitativa ordinal
* Filhos - Quantitativa discreta
* Salario -	Quantitativa contínua
* Anos (na empresa)	- Quantitativa discreta

## Biblioteca Gráfica Seaborn


````
import matplotlib.pyplot as plt 
# Importando seaborn para gráficos mais complexos e elaborados
import seaborn as sns
````

## Comparação com Pandas
Para fins de comparação com o Seaborn vamos relembra um gráfico de barras produzido com o `plot()` do Pandas.

````
# Mostrar a quantidade de funcionários por nível de instrução
milsa.groupby("Instrucao")["Instrucao"].count().plot(kind="bar")
````

## Gráfico de Barras
Note abaixo que a contagem (que acima foi feita com groupby) é feita por meio do parâmetro `kind = "count"`.

````
# Com Seaborn
sns.factorplot('Instrucao', data=milsa, kind='count')
````

## Gráfico de Dispersão
Objetivo: Visualizar a relação entre as "Anos de Empresa" e "Salário" colorindo cada ponto com a variável "Instrução".

````
sns.relplot(x="Anos", y="Salario", hue="Instrucao", data=milsa)
````

## Gráfico Boxplot
Objetivo: Visualizar o boxplot do `Salário` por Grau de `Instrução`.

````
sns.factorplot(x="Instrucao", y="Salario", kind="box", data=milsa) 
````

## Gráfico Strip
Objetivo: Visualizar na forma de Strip (tiras) o `Salário` por Grau de `Instrução`.

````
sns.factorplot(x="Instrucao", y="Salario", kind="strip", data=milsa) 
````

## Gráfico de Barras com 3 variáveis

Objetivo: Visualizar o `Salário` por Grau de `Instrução` e `Região`.

````
sns.factorplot(x ='Regiao', y ='Salario',  
               kind = 'bar', hue = 'Instrucao',  
               data = milsa) 
````
