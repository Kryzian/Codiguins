## Seleção Simples (if)

A `seleção simples` tem a seguinte estrutura:

```
if condicao:
  comandos
  comandos
  comandos
```

[Observações]:
* A "condição" é uma expressão que "compara" pelo menos dois valores com operadores relacionais (e/ou lógicos). 
> Exemplo 1: a > b
* Após a "condição" no ``if`` temos ``:`` (dois-pontos)
* Os "comandos" logo abaixo do "if" são executados se o teste de "condição" der um resultado "verdadeiro" (true)
* Os "comandos" que estão "dentro do if" são identificados pelo "deslocamento à direita" (espaços no começo da linha) em relação ao "if"
> * Estes deslocamentos são chamados de "indentação" (ou endentação)
> * Em Python o uso de indentação é obrigatório,  pois é isto que identifica o que está "dentro" ou "fora" de um "bloco".
> * Recomenda-se usar o TAB para fazer a indentação, apesar se ser possível usar também a "barra de espaço" do teclado.

### Solução do Problema 2

Para veficar se a correlação é negativa a condição a ser utilizada é `"cor < 0"` 

````
# Solução Problema [2]
coluna = input("Digite a coluna: ")
cor = pac['IMC'].corr(pac[coluna])
print("Correlação entre IMC e", coluna, ": %.4f" % cor)
if  (cor < 0) :
  print("Atenção: Correlação negativa!")
````

A `seleção composta` tem a seguinte estrutura:


Somente se a moda tiver dois valores deve-se exibir a mensagem "Bimodal".

[Atenção]: 
* mode() retorna a moda
* mode().count() retorna a quantidade de elementos "da moda".

### Solução do Problema 4

````
# Solução Problema [4]
coluna = input("Digite a coluna: ")
moda = pac[coluna].mode()
print("moda do atributo ", coluna, " = ", moda)
if  (moda.count() == 2) :
  print("Bimodal!")
````

Execute a solução acima 2 vezes:

* Informe Idade na primeira execução
* Informe Pulsacao na segunda execução

Note que:
* Na primeira execução (Idade) não é exibida mensagem adicional, pois a quantidade da moda é 1.
* Na segunda execução (Pulsacao) a mensagem exibida foi "Bimodal!"

Note que no caso de "Pulsacao" há dois valores: 64 (na posição 0) e 72 (na posição 1).

#### Inibindo a impressão do índice e header
[ATENÇÃO:] Para NÃO imprimir os índices (0, 1, ...) e o "dtype" podemos utilizar o `to_string()` com os parâmetros `index=False` e/ou `header=False`. Vide exemplo abaixo

````
#df.to_string(index=False, header=False)
# Solução Problema [4]
coluna = input("Digite a coluna: ")
moda = pac[coluna].mode()
print("moda do atributo ", coluna, "=", moda.to_string(index=False))
if  (moda.count() == 2) :
  print("Bimodal!")
````
