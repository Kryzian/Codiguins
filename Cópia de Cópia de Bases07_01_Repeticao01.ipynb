## Estrutura de Repetição - Enquanto

A estrutura de repetição enquanto permite especificar instruções que devem ser repetidas enquanto determinada condição for verdadeira.

A estrutura de repetição  `enquanto` `(while)` tem a seguinte estrutura:

```
while condicao:
  comandos
  comandos
  comandos
```

[Observações]:
* A "condição" é uma expressão que "compara" pelo menos dois valores com operadores relacionais (e/ou lógicos). 
> Exemplo 1: a > b
* Após a "condição" no ``while`` temos ``:`` (dois-pontos)
* Os "comandos" logo abaixo do "while" são executados se o teste de "condição" der um resultado "verdadeiro" (true)
* Os "comandos" que estão "dentro do while" são identificados pelo "deslocamento à direita" (espaços no começo da linha) em relação ao "while"
> * Estes deslocamentos são chamados de "indentação" (ou endentação)
> * Em Python o uso de indentação é obrigatório,  pois é isto que identifica o que está "dentro" ou "fora" de um "bloco".
> * Recomenda-se usar o TAB para fazer a indentação, apesar se ser possível usar também a "barra de espaço" do teclado.

### Solução do Problema 2 (com repetição)

A solução para o problema #2 usando estrutura de repetição enquanto é apresentada a seguir. Note que a solução é bem curta se comparada a solução "passo a passo" (3 linhas repetidas 11 vezes).

````
# com repetição
x = mediana - 1
limite = mediana + 1.1
while (x <= limite):
  y = m*x+b  
  print("A predição para Peso = %.2f" % x, " é IMC = %.2f" % y)
  x = x + 0.2
````

### Solução Completa

````
#solução completa
import numpy as np
m, b = np.polyfit(x=pac['Peso'], y=pac['IMC'], deg = 1)
print("equação => y=%.2f " %m, "*x + %.2f" %b)
mediana = pac.Peso.median()
print("mediana:", mediana)
#------------------------------- repetição
x = mediana - 1
limite = mediana + 1.1
while (x <= limite):
  y = m*x+b  
  print("A predição para Peso = %.2f" % x, " é IMC = %.2f" % y)
  x = x + 0.2
````
