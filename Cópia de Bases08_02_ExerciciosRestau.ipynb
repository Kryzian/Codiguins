## 1) Problema 1

Escreva um programa para exibir:
* Quantas contas foram pagas por Sexo
* A média das contas por Sexo
* A média das gorgetas por Sexo

```
#sua solução usando iteração

import pandas as pd 
#https://drive.google.com/file/d/1o8hOIzSbFeBMue5iM_x8dl26mLfrkHbh/view?usp=sharing
codigo = "1o8hOIzSbFeBMue5iM_x8dl26mLfrkHbh"
file = "https://drive.google.com/u/3/uc?id=" + codigo + "&export=download"
restau = pd.read_csv(file, sep = ";")
cont1=0
cont2=0
sid1=0
sid2=0
gorj1 = 0
gorj2 = 0

for ind, cli in restau.iterrows():
  if cli.Sexo == "Masculino":
    cont1=cont1+1
    sid1=sid1+cli.TotalConta
    gorj1 = gorj1+ cli.Gorjeta
  else:
    cont2 = cont2+1
    sid2=sid2+cli.TotalConta
    gorj2 = gorj2+ cli.Gorjeta
print("Masculino :", cont1, "Feminino:", cont2)
if (cont1>0):
  contmed1 = sid1/cont1
  gorjmed1 = gorj1/cont1
  print("Conta Media - Masculino:", "%.2f" %  contmed1)
  print("Gorjeta Media - Masculino:", "%.2f" % gorjmed1)
if (cont2>0):
  contmed2 = sid2/cont2
  gorjmed2 = gorj2/cont2
  print("Conta Media - Feminino:", "%.2f" % contmed2)
  print("Gorjeta Media - Feminino:", "%.2f" % gorjmed2)
```

## 2) Problema 2

Escreva um programa para exibir:
* Quantos pratos por tamanho pequeno[1 ou 2], médio[3 ou 4] ou grande [5 ou 6] 
* A soma das contas por tamanho do prato: pequeno, médio ou grande
* A soma das gorgetas por tamanho do prato: pequeno, médio ou grande 

[ATENÇÃO]: Lembre-se de usar `parênteses` com o operador lógico "ou" (|). 

Exemplo:  `if` **(**x `==`1**)** `|` **(**x `==` 2**)**:

```
# sua solução com iteração

import pandas as pd 
#https://drive.google.com/file/d/1o8hOIzSbFeBMue5iM_x8dl26mLfrkHbh/view?usp=sharing
codigo = "1o8hOIzSbFeBMue5iM_x8dl26mLfrkHbh"
file = "https://drive.google.com/u/3/uc?id=" + codigo + "&export=download"
restau = pd.read_csv(file, sep = ";")
cont1=0
cont2=0
cont3 = 0
sid1=0
sid2=0
sid3 = 0
gorj1 = 0
gorj2 = 0
gorj3 = 0

for ind, tam in restau.iterrows():
  if (tam.Tamanho == 1) | (tam.Tamanho == 2):
    cont1 = cont1 +1
    sid1 = tam.TotalConta + sid1
    gorj1 = gorj1 + tam.Gorjeta
  elif (tam.Tamanho == 3) | (tam.Tamanho == 4):
    cont2 = cont2 +1
    sid2 = tam.TotalConta + sid2
    gorj2 = gorj2 + tam.Gorjeta
  else:
    cont3 = cont3 + 1
    sid3 = tam.TotalConta + sid3
    gorj3 = gorj3 + tam.Gorjeta
print("Pequeno:",cont1,"Médio:", cont2,"Grande:",cont3)
if cont1 >0 :
  contmed1 = sid1/cont1
  gorjmed1 = gorj1/cont1
  print("Conta Média - Pequeno:", "%.2f" %  contmed1)
  print("Gorjeta Média - Pequeno:", "%.2f" % gorjmed1)
if (cont2>0):
  contmed2 = sid2/cont2
  gorjmed2 = gorj2/cont2
  print("Conta Média - Médio:", "%.2f" % contmed2)
  print("Gorjeta Média - Médio:", "%.2f" % gorjmed2)
if (cont3>0):
  contmed3 = sid3/cont3
  gorjmed3 = gorj3/cont3
  print("Conta Média - Grande:", "%.2f" % contmed3)
  print("Gorjeta Média - Grande:", "%.2f" % gorjmed3)
```
