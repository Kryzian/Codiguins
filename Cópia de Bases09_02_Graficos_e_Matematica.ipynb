Neste caderno iremos explorar problemas que envolvem os pontos de intersecção entre funções. 

Os pontos de intersecção entre duas funções são os valores para x e y que satisfazem simultaneamente as duas funções.

## Importando bibliotecas

Para nossos exemplos e exercícios vamos fazer uso de duas biblitecas básicas. Uma para "plotar gráficos" (`pyplot`) e outra para "manipular números" (`numpy`)

````
#importando a biblioteca numpy
import numpy as np
#importando a biblioteca para gráficos
import matplotlib.pyplot as plt
````

## Exemplo 1

Os exemplos 1 e 2 baseiam-se no material disponível [neste link](https://mundoeducacao.uol.com.br/matematica/pontos-interseccao-entre-funcoes.htm).

Dadas as funções f = x + 1 e g = 2x – 1,  vamos calcular o ponto de intersecção das funções.

Temos duas funções do 1º grau. Vale lembrar que a representação de uma função do 1º grau no plano cartesiano é uma reta.

Como as duas funções são simples podemos rapidamente encontrar os valores de x, para os quais y = 0.
* Para f(x) : (i)  x+1=0;  (ii) x=-1 
* Para g(x) : (i)  2x-1=0; (ii) 2x=1;  (iii) x = 1/2

Ou seja, x=-1 e x=1/2 são pontos em que cada gráfico "toca" o ponto y=0

### Plotando o gráfico
Vamos inicialmente plotar um gráfico, para as duas funçoes abrangendo esses dois pontos de x (-1 e 1/2). Por exemplo, o intervalo de -3 a 3.

Como as duas funções são de primeiro grau vamos usar intervalo inteiro (igual as 1)

````
# 0)Aumentando o tamanho do gráfico
plt.rcParams['figure.figsize'] = (10,6)
# 1) Gerando uma série de números variando de 1 em 1
inicio=-3
fim = 3
x = np.arange(inicio, fim, 1)  
# 2) calculando os valores de f(x) e g(x)
f=x+1
plt.plot(x, f,linewidth=2.0, linestyle="-", color="blue")
g=2*x-1
plt.plot(x, g,linewidth=2.0, linestyle="-", color="green")
# Capturando os elementos da região da figura
ax = plt.gca()
# Setando os valores do eixo x de meio em meio
ax.set_xticks(np.arange(inicio, fim, 0.5))
# Colocando um grid (grade) no gráfico
plt.grid()
# adicionando as legendas - loc=0 faz com que localize o melhor local para legenda
plt.legend(['f(x)=x+1', 'g(x) =2x-1'], loc=0)
# --- alterando as linhas de delimitação do gráfico
#a) Retirar a linha da DIREITA (color = none)
ax.spines['right'].set_color('none')
#b) Retirar a linha de CIMA (color = none)
ax.spines['top'].set_color('none')
#c) Posicionar a linha de BAIXO, exatamente no "y=0"
ax.spines['bottom'].set_position(('data',0))
#d) Posicionar a linha da ESQUERDA, exatamente no "x=0"
ax.spines['left'].set_position(('data',0))
#------
# Mostrando o gráfico
plt.show()
````

Notamos que os dois gráficos se encontram nas proximidades de x=2. Vamos então mudar o intervalo do gráfico para -2 a 4 (ao invés de -3 a 3). Vide abaixo.

````
# 0)Aumentando o tamanho do gráfico
plt.rcParams['figure.figsize'] = (10,6)
# 1) Gerando uma série de números variando de 1 em 1
inicio=-2
fim = 4
x = np.arange(inicio, fim, 1)  
# 2) calculando os valores de f(x) e g(x)
f=x+1
plt.plot(x, f,linewidth=2.0, linestyle="-", color="blue")
g=2*x-1
plt.plot(x, g,linewidth=2.0, linestyle="-", color="green")
# Capturando os elementos da região da figura
ax = plt.gca()
# Setando os valores do eixo x de meio em meio
ax.set_xticks(np.arange(inicio, fim, 0.5))
# Colocando um grid (grade) no gráfico
plt.grid()
# adicionando as legendas - loc=0 faz com que localize o melhor local para legenda
plt.legend(['f(x)=x+1', 'g(x) =2x-1'], loc=0)
# --- alterando as linhas de delimitação do gráfico
#a) Retirar a linha da DIREITA (color = none)
ax.spines['right'].set_color('none')
#b) Retirar a linha de CIMA (color = none)
ax.spines['top'].set_color('none')
#c) Posicionar a linha de BAIXO, exatamente no "y=0"
ax.spines['bottom'].set_position(('data',0))
#d) Posicionar a linha da ESQUERDA, exatamente no "x=0"
ax.spines['left'].set_position(('data',0))
#------
# Mostrando o gráfico
plt.show()
````

### Identificando o ponto de intersecção
Podemos aplicar o x=2 nas duas funções para verificar os valores. Vide abaixo.


````
x=2
f=x+1
g=2*x-1
print("f(", x, ") =", f)
print("g(", x, ") =", g)
````

Note que f(2) e g(2) têm o mesmo valor (3). Assim, dizemos que o ponto de intersecção das retas de y = x + 1 e y = 2x – 1 é o ponto que possui coordenadas (2, 3).

### Onde os gráficos se cruzam?

Para melhor identificar estes pontos podemos instruir que seja exibido os pontos (valores de x) cuja diferença entre f(x) e g(x) seja próxima de zero. Por exemplo, menor que 0.01.

````
x = np.arange(-2,4)  
f=x+1
g=2*x-1
for i in range(f.size):  
    difer = abs(f[i] - g[i])   # calcula a diferença entre f(x) e g(x) em valor absoluto
    # verifica se essa diferença é menor que 0.01
    if (difer < 0.01):
       print("x=",x[i], ": f(x) = %.4f " % f[i], " e g(x) = %.4f" % g[i], "diferença:%.4f" % difer)
````

Note que o código acima exibiu exatamente o ponto x=2. A diferença igual a zero indica que o ponto identificado é bem preciso. Isto ocorre porque estamos lidando com funções de primeiro grau. Vejamos outro exemplo envolvendo funções um pouco mais complexas.


## Exemplo 2

Dadas as funções  $f = 2x$ e $g = – x^2 + 4x$ , vamos calcular o ponto de intersecção das funções.

Temos uma função de 1º grau e outra de  2º grau.

Vamos encontar para a função de  1º grau o valor de x em que y = 0. Isso vai ajudar a enxergar um intervalo para plotar o gráfico

Para f(x) : (i) 2x=0; (ii) x=0/2;  (iii) x = 0


### Plotando o gráfico
Vamos inicialmente plotar um gráfico, para as duas funçoes passando por x=0. Por exemplo, o intervalo de -3 a 3.

Como há uma função de 2º grau vamos usar um intervalo não inteiro (0.01).

````
# 0)Aumentando o tamanho do gráfico
plt.rcParams['figure.figsize'] = (10,6)
# 1) Gerando uma série de números 
inicio=-3
fim = 3
x = np.arange(inicio, fim, 0.01)  
# 2) calculando os valores de f(x) e g(x)
f=2*x
plt.plot(x, f,linewidth=2.0, linestyle="-", color="blue")
g=-x**2+4*x
plt.plot(x, g,linewidth=2.0, linestyle="-", color="green")
# Capturando os elementos da região da figura
ax = plt.gca()
# Setando os valores do eixo x de meio em meio
ax.set_xticks(np.arange(inicio, fim, 0.5))
# Colocando um grid (grade) no gráfico
plt.grid()
# adicionando as legendas - loc=0 faz com que localize o melhor local para legenda
plt.legend(['f(x)=2x', 'g(x) =-x^2 + 4x'], loc=0)
# --- alterando as linhas de delimitação do gráfico
#a) Retirar a linha da DIREITA (color = none)
ax.spines['right'].set_color('none')
#b) Retirar a linha de CIMA (color = none)
ax.spines['top'].set_color('none')
#c) Posicionar a linha de BAIXO, exatamente no "y=0"
ax.spines['bottom'].set_position(('data',0))
#d) Posicionar a linha da ESQUERDA, exatamente no "x=0"
ax.spines['left'].set_position(('data',0))
#------
# Mostrando o gráfico
plt.show()
````

Notamos que os dois gráficos se encontram nas proximidades de x=0 e x=2. Vamos então mudar o intervalo do gráfico para -1 a 5 (ao invés de -3 a 3) para tentar ver a parábola completa da função g(x). Vide abaixo.

````
# 0)Aumentando o tamanho do gráfico
plt.rcParams['figure.figsize'] = (10,6)
# 1) Gerando uma série de números 
inicio=-1
fim = 5
x = np.arange(inicio, fim, 0.01)  
# 2) calculando os valores de f(x) e g(x)
f=2*x
plt.plot(x, f,linewidth=2.0, linestyle="-", color="blue")
g=-x**2+4*x
plt.plot(x, g,linewidth=2.0, linestyle="-", color="green")
# Capturando os elementos da região da figura
ax = plt.gca()
# Setando os valores do eixo x de meio em meio
ax.set_xticks(np.arange(inicio, fim, 0.5))
# Colocando um grid (grade) no gráfico
plt.grid()
# adicionando as legendas - loc=0 faz com que localize o melhor local para legenda
plt.legend(['f(x)=2x', 'g(x) =-x^2 + 4x'], loc=0)
# --- alterando as linhas de delimitação do gráfico
#a) Retirar a linha da DIREITA (color = none)
ax.spines['right'].set_color('none')
#b) Retirar a linha de CIMA (color = none)
ax.spines['top'].set_color('none')
#c) Posicionar a linha de BAIXO, exatamente no "y=0"
ax.spines['bottom'].set_position(('data',0))
#d) Posicionar a linha da ESQUERDA, exatamente no "x=0"
ax.spines['left'].set_position(('data',0))
#------
# Mostrando o gráfico
plt.show()
````

### Onde os gráficos se cruzam?

Para melhor identificar estes pontos podemos instruir que seja exibido os pontos (valores de x) cuja diferença entre f(x) e g(x) seja próxima de zero. Por exemplo, menor que 0.01.

````
x = np.arange(-1,5,0.01)  
f=2*x
g=-x**2+4*x
for i in range(f.size):  
    difer = abs(f[i] - g[i])   # calcula a diferença entre f(x) e g(x) em valor absoluto
    # verifica se essa diferença é menor que 0.01
    if (difer < 0.01):
       print("x=%.4f" % x[i], ": f(x) = %.4f " % f[i], " e g(x) = %.4f" % g[i], "diferença:%.4f" % difer)
````

Note que o código acima exibiu os valores x=0 e x=2. 

## Exemplo 3

* $f(x) = x^3$
* $g(x) = 9x - 3$




Para g(x) temos que:
* (i) 9x-3=0;  (ii) 9x=3;  (iii) x=3/9; (iv) x = 1/3

Vamos inicialmente plotar um gráfico, para as duas funçoes passando por x=1/3. Por exemplo, o intervalo de -3 a 3.

Como há uma função de 3º grau vamos usar um intervalo não inteiro (0.001).

### Plotando o gráfico

````
# Aumentando o tamanho do gráfico
plt.rcParams['figure.figsize'] = (11,7)
# 1) Gerando uma série de números variando de 0.001
inicio=-3
fim = 3
x = np.arange(inicio, fim, 0.001)
# 2) calculando os valores de g1(x) e g2(x)
f = x**3
plt.plot(x, f,linewidth=2.0, linestyle="-", color="blue")
g = 9*x - 3
plt.plot(x, g, linewidth=2.0, linestyle="-", color="green")
# Capturando os elementos da região da figura
ax = plt.gca()
# Setando os valores do eixo x de meio em meio
ax.set_xticks(np.arange(inicio, fim, 0.5))
# Colocando um grid (grade) no gráfico
plt.grid()
# adicionando as legendas - loc=0 faz com que localize o melhor local para legenda
plt.legend(['f(x)=x^3', 'g(x) = 9x-3'], loc=0)
# --- alterando as linhas de delimitação do gráfico
#a) Retirar a linha da DIREITA (color = none)
ax.spines['right'].set_color('none')
#b) Retirar a linha de CIMA (color = none)
ax.spines['top'].set_color('none')
#c) Posicionar a linha de BAIXO, exatamente no "y=0"
ax.spines['bottom'].set_position(('data',0))
#d) Posicionar a linha da ESQUERDA, exatamente no "x=0"
ax.spines['left'].set_position(('data',0))
#------
# Mostrando o gráfico
plt.show()
````

### Onde os gráficos se cruzam?

Para melhor identificar estes pontos podemos instruir que seja exibido os pontos (valores de x) cuja diferença entre f(x) e g(x) seja próxima de zero. Por exemplo, menor que 0.01.

````
x = np.arange(-3,3,0.001)  
f = x**3
g = 9*x - 3
for i in range(f.size):  
    difer = abs(f[i] - g[i])   # calcula a diferença entre f(x) e g(x) em valor absoluto
    # verifica se essa diferença é menor que 0.01
    if (difer < 0.01):
       print("i=", i, "- x=%.4f" % x[i], ": f(x) = %.4f " % f[i], " e g(x) = %.4f" % g[i], "diferença:%.4f" % difer)
````

Note que o código acima exibiu os valores x=0.3380 e x=2.8170 como as duas menores diferenças (0.0034 e 0.0013). São respectivamente as posições 3338 e 5817 no conjunto x. Os dois valores de x coincidem (aproximadamente) com o cruzamento das funções na imagem acima.
