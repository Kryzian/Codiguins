## Estrutura de repetição for
A estrutura de repetição  `para` `(for)` tem a seguinte estrutura:

```
for elemento  in conjunto:
  comandos
  comandos
  comandos
```

[Observações]:
* `elemento` é um item do `conjunto`.
* `conjunto` pode ser uma tabela do pandas ou outro tipo de conjunto (uma lista, por exemplo).
* Após o conjunto no ``for`` temos ``:`` (dois-pontos).
* Os "comandos" logo abaixo do "for" são executados até que todo o conjunto seja percorrido ou haja uma instrução de parada.
* **IMPORTANTE**: no `for` há uma instrução "`next`" (próximo) implicita, ou seja, no final do for, um `next` é executado automaticamente para avançar para o próximo elemento (vide imagem após solução completa do problema 1).
* Os "comandos" que estão "dentro do for" são identificados pelo "deslocamento à direita" (espaços no começo da linha) em relação ao "for"
> * Estes deslocamentos são chamados de "indentação" (ou endentação)
> * Em Python o uso de indentação é obrigatório,  pois é isto que identifica o que está "dentro" ou "fora" de um "bloco".
> * Recomenda-se usar o TAB para fazer a indentação, apesar se ser possível usar também a "barra de espaço" do teclado.

Há diversas "variações" da **estrutura de repetição for**. 

Para o uso com uma tabela acessada via pandas uma forma bastante útil é com a função iterrows(). 

A função iterrrows() "itera" (repete) para cada elemento da tabela, retornando dois valores:
* o primeiro valor é o índice (um número correspondendo ao número da linha na tabela)
* o segundo valor corresponde aos dados da respectiva linha.

Vejamos a solução para o Problema 1 como exemplo.

### Solução do Problema 1 (completa)

````
media = pac.Idade.mean()
print("Média das idades: %.2f" % media)
# usa iterrows to obter os dados de cada linha do dataframe
for indice, valor in pac.iterrows(): 
    if valor.Idade > media:
      print("Paciente:", indice, "Idade: ",valor.Idade, " - Acima da Média")
    else:
      if valor.Idade < media:
        print("Paciente:", indice, "Idade: ",valor.Idade, " - Abaixo da Média")
      else:
        print("Paciente:", indice, "Idade: ",valor.Idade, " - Exatamente na Média")
    
````

## Problema 2

Incremente a solução anterior para, ao final, exibir também:
* Quantidade de pacientes acima da média
* Quantidade de pacientes abaixo da média
* Quantidade de pacientes exatamente na média

Para este incremento, vamos precisar utilizar 3 contadores (acumuladores).

Um “contador” tem a seguinte forma:
* var = var + 1    

ou seja  a variável “recebe” o valor que ela já possui  somado com mais 1.

Um contador, deve ser inicializado com o início da contagem. Normalmente, começa-se a contar do **0** (zero) do **1** (um).


### Solução completa Problema 2

````
# solução completa - mensagens e contadores
media = pac.Idade.mean()
print("Média das idades: %.2f" % media)
# usa iterrows to obter os dados de cada linha do dataframe
cont1=0  
cont2=0
cont3=0
for indice, valor in pac.iterrows(): 
    if valor.Idade > media:
      print("Paciente:", indice, "Idade: ",valor.Idade, " - Acima da Média")
      cont1 = cont1 + 1
    else:
      if valor.Idade < media:
        print("Paciente:", indice, "Idade: ",valor.Idade, " - Abaixo da Média")
        cont2 = cont2 + 1
      else:
        print("Paciente:", indice, "Idade: ",valor.Idade, " - Exatamente na Média")
        cont3 = cont3 + 1
#-- mostra os contadores
print("Acima da média", cont1)
print("Abaixo da média", cont2)
print("Exatamente na média", cont3)
````

Veja como o "`next`" está implícito no "`for`".

### Solução Problema 2 (somente os contadores)

Vejamos a solução do problema 2, exibindo apenas os totais (contadores)

````
# exibindo somente os contadores
media = pac.Idade.mean()
print("Média das idades: %.2f" % media)
# usa iterrows to obter os dados de cada linha do dataframe
cont1=0  
cont2=0
cont3=0
for indice, valor in pac.iterrows(): 
    if valor.Idade > media:      
      cont1 = cont1 + 1
    elif valor.Idade < media:
      cont2 = cont2 + 1
    else:
      cont3 = cont3 + 1
#-- mostra os contadores
print("Acima da média", cont1)
print("Abaixo da média", cont2)
print("Exatamente na média", cont3)
````

### Mensagens dos contadores "dentro do for"

Veja o que acontece se cada "print" que mostra os contadores estiver "dentro do for".

````
media = pac.Idade.mean()
print("Média das idades: %.2f" % media)
# usa iterrows to obter os dados de cada linha do dataframe
cont1=0  
cont2=0
cont3=0
for indice, valor in pac.iterrows(): 
    if valor.Idade > media:
      cont1 = cont1 + 1
    else:
      if valor.Idade < media:
        cont2 = cont2 + 1
      else:
        cont3 = cont3 + 1
    #-- mostra os contadores
    print("Acima da média", cont1)
    print("Abaixo da média", cont2)
    print("Exatamente na média", cont3)
````

Note que cada "trio de mensagens" foi exibido para cada "paciente". Isto não está adequado pois o enunciado diz que é para exibir "ao final".
