Gráficos são recursos visuais muito utilizados para facilitar a leitura e a compreensão de informações sobre fenômenos e processos naturais, sociais e econômicos, etc.

No cotidiano, jornais, revistas e livros, além de programas educativos, mostram o quanto esse recurso é explorado pelos meios de comunicação.

Gráficos são muito úteis para **expressar visualmente** os dados ou valores obtidos por meio da estatística descritiva (médias, desvios, etc), de modo a facilitar a compreensão.
 


## Tipos de Gráficos

Existem vários tipos de gráficos: de barra (horizontal ou vertical), de linha, de dispersão, de setor (pizza), etc.

O tipo de gráfico mais apropriado depende dos tipos das variáveis a serem exibidas no gráfico e também do seu papel (dependente ou independente).


## Tipos de Variáveis

Para uma análise descritiva adequada é importante verificar os tipos variáveis disponíveis no conjunto. 

Devemos lembrar que cada uma das características dos dados da população (ou amostra) é denominada de variável. Por exemplo: nível de instrução, anos de trabalhos na empresa, salário, etc.
As variáveis podem assumir diferentes valores, que basicamente podem ser separados em
*	Quantitativos ou numéricos
*	Qualitativos ou não numéricos ou categóricos

## Classificação de variáveis

As variáveis quantitativas ou numéricas podem ser
*	Discretas: assumem apenas valores inteiros. Ex.: quantidade de filhos
*	Contínuas: assume qualquer valor no intervalo dos números reais. Ex.: salário

As variáveis qualitativas ou categóricas podem ser
*	Nominais: quando as categorias não possuem uma ordem natural. Ex.: estado civil.
*	Ordinais: quando as categorias podem ser ordenadas. Ex.: nível de escolaridade (1o grau, 2o grau, Superior).

## Gráficos com Pandas

Pandas facilita o trabalho de criar os gráficos encapsulando a biblioteca `Matplotlib`.

Para criar os gráficos precisamos apenas chamar a função `plot()` para toda a tabela ou para atributos (colunas). Por padrão o Pandas irá exibir os dados como um gráfico de linha.

Por exemplo:

````
milsa.plot()
````

Mas o gráfico de linha a acima não faz muito sentido, pois mistura todos os atributos que estão em unidades de medida distintas.

Vejamos o gráfico de apenas um atributo, por exemplo, `Anos`.

````
milsa.Anos.plot()
````

O gráfico acima mostra um gráfico de linhas, apenas do atributo Idade (eixo Y). O eixo X é a primeira coluna da tabela, ou seja, o número do funcionário (1, 2, 3, etc). O gráfico não diz muito pois esta variação (subidas e descidas) em relação ao "número do funcionário" (identificação) não faz muito sentido.

Podemos, por exemplo, ordenar os dados de `Anos`. Veja o que acontece.

````
milsa.Anos.sort_values().plot()
````

O gráfico não melhorou muito. Ele apenas parte do menos valor de `Anos na Empresa` e vai crescendo, sempre associado ao atributo "Funcionario".  


A conclusão que devemos tirar dessas análises preliminares, com gráficos, é que é muito importante escolher bem os dados a serem exibidos, ou seja, planejar bem a consulta. Devemos partir da consulta (pergunta), entender bem o resultado da consulta e então plotar o gráfico.

## Exemplo [1]
Objetivo: Analisar a relação entre Salário e Tempo na Empresa (em Anos)

Com o `plot()` podemos especificar os parâmetros x e y, para indicar que "atributo" dever ser utilizado em cada eixo.

### Gráfico de Linhas

````
milsa.plot(x='Anos', y= 'Salario')
````

Vejamos como fica o gráfico de linhas, com os mesmos dados, mas ordenados por `Anos` de Empresa.

````
milsa.sort_values(by="Anos").plot(x='Anos', y= 'Salario')
````

O gráfico de linhas acima mostra muitas váriações, subidas e descidas, em relação ao eixo X, indicando não haver "uma relação linear" entre as duas variáveis.

### Gráfico de Barras

Vejamos um gráfico de barras. Basta usar o parâmetro kind = "bar" conforme abaixo.



````
milsa.plot(x='Anos', y= 'Salario',kind="bar")
````

Vejamos como fica o gráfico de barras, com os mesmos dados, mas ordenados por Anos de Empresa.

````
milsa.sort_values(by="Anos").plot(x='Anos', y= 'Salario', kind = "bar")
````

Abaixo um gráfico de barras apenas considerando a coluna "Anos". O eixo "x" neste caso é a identicação do funcionário.

````
milsa.Anos.plot(kind="bar")
````

### Gráfico de Dispersão

Vejamos um gráfico de barras. Basta usar o parâmetro kind = "scatter" conforme abaixo.

````
milsa.plot(x='Anos', y= 'Salario', kind='scatter')
````

Vejamos como fica o gráfico de dispersão, com os mesmos dados, mas ordenados por Anos de Empresa.

````
milsa.sort_values(by="Anos").plot(x='Anos', y= 'Salario', kind='scatter')
````

Apenas por curiosidade vamos dar uma olhada no coeficiente de correlação?

````
milsa['Anos'].sort_values().corr(milsa['Salario'])
````

Esta correlação de 0.36 é considerada **FRACA**.

### Gráfico de Setor (Pizza ou Torta)

Para construir um gráfico de setor, chamado de pizza ou torta (pie), basta usar o parâmetro kind = "pie".

Mas um gráfico do tipo "pie" é recomendado para variáveis qualitativas (nominais ou ordinais). As seguintes colunas de "milsa" são qualitativas: 
* EstCivil
* Instrucao
* Regiao

Objetivo: Exibir graficamente a **quantidade** de funcionários por nível de `Instrução` (**nominal ou ordinal**)


````
# Quantos funcionarios existem por nível de instrução?
QInstrucao = milsa.groupby("Instrucao")["Instrucao"].count()
QInstrucao
````

````
QInstrucao.plot(kind="pie")
````

Poderíamos usar diretamente em uma única instrução. Mas a instrução fica mais complexa.

````
milsa.groupby("Instrucao")["Instrucao"].count().plot(kind="pie")
````

### Gráfico de Barras (Nominal)

Objetivo: Exibir graficamente a média dos funcionários por nível de `Instrução` (**nominal ou ordinal**)

````
# Qual a média dos salário por nível de instrução?
MSalario = milsa.groupby("Instrucao")["Salario"].mean()
MSalario
````

````
MSalario = milsa.groupby("Instrucao")["Salario"].mean()
MSalario.plot(kind="bar")
````

Poderíamos usar diretamente em uma única instrução. Mas a instrução fica mais complexa.

````
milsa.groupby("Instrucao")["Salario"].mean().plot(kind="bar")
````

### Gráfico de Barras Horizontal

Vejamos um gráfico de barras HOZIONTAL. Basta usar o parâmetro kind = "barh" conforme abaixo.

````
milsa.groupby("Instrucao")["Salario"].mean().plot(kind="barh")
````

### Gráfico - Histograma

Histogramas são gráficos similares (mas não iguais) aos gráficos de barras e são associados a frequência de valores de um (ou mais) atributo(s).

**Objetivo**: Qual a distribuição (quantitativa) dos funcionários da empresa Milsa, por Ano (ou faixa de anos)?

````
milsa.Anos.plot(kind="hist")
````

O histograma acima exibe na base (eixo x) os "Anos" e em cada retângulo (altura) a área proporcional às frequências desta classe (Anos).

Ao especificar o tipo (kind = hist) o gráfico é montado a partir de uma contagem (frequência), utilizando "10 caixas" (10 bins), ou seja 10 retângulos.

Podemos especificar o número de caixas,  por meio do parâmetro "bins".

````
# 5 bins
milsa.Anos.plot(kind="hist", bins=5)
````

````
# 20 bins
milsa.Anos.plot(kind="hist", bins=20)
````

É importante registra que o "hist" só funciona para variáveis numéricas, ou seja, a "distribuição" em faixa de valores numéricos.

Para dados não númericos, como "Instrução" construimos o gráfico de barras, a partir de uma contagem prévia.

Para mais explicações sobre a diferença entre gráfico de barras e histograma acesse [este link](http://www.shodor.org/interactivate/discussions/HistogramsVsBarGraph/) (em inglês)

### Outros parâmetros

Independente do tipo de gráfico, podemos alterar a cor (color), acrescentar título ao gráfico (title), alterar o tamanho da imagem (figsize), etc. Vide abaixo.

````
import pandas as pd
codigo = "1_c1rtxf2YDaBKeQA62Gg4yjljuVO6mgc"
file = "https://drive.google.com/u/3/uc?id=" + codigo + "&export=download"
milsa = pd.read_csv(file, sep = ";",decimal=",")
milsa.Anos.plot(kind="hist", bins = 20, color='purple', title="Funcionários x Anos de Empresa", figsize= (10, 6))
````

Uma forma de alterar legendas e título é usar funções como set_title, set_xlabel, etc.

Mas antes disso, precisamos armazenar em uma variável (por exemplo, ax) o restultado de plot(). Depois disso, podem usar "ax" (ou outro nome de variável) para "chamar" estas outras funções.  Vide abaixo.

````
ax =milsa.Anos.plot(kind="hist", color='black')
ax.set_title("Funcionários x Anos de Empresa")
ax.set_xlabel("Anos Empresa")
ax.set_ylabel("Quantidade Funcionários")
````

### Gráfico Boxplot

Gráficos do tipo "box plot" (diagrama de caixa), são muito úteis para representar graficamente as medidas de tendência central e dispersão.

Para construir um gráfico do tipo `Boxplot` basta usar o parâmetro kind = "box".

````
milsa.plot(kind="box")
````

O parâmetro "rot = 90" faz com que os rótulos (Filhos, Salario, etc) girem 90 graus (vertical). Vide abaixo

````
# rot = 90 
milsa.plot(kind="box", rot=90)
````

O gráfico acima mostra um "box" para cada variável numérica. Como as unidades de medida de cada atributo são distintas não faz muito sentido ter todas elas no mesmo gráfico. Vamos plotar um gráfico do tipo box para a variável "Filhos" para facilitar a análise.

````
milsa.Filhos.plot(kind="box")
````

No box plot:
* a linha verde dentro da caixa correponde a mediana. 
* a "caixa" corresponde ao 1o. e 3o. quartis (vide explicações abaixo).
* A barra que acompanha a caixa corresponde a amplitude entre o 1o e 3o. quartil.
* Valores fora dessa amplitude normalmente são considerados como discrepantes com relação à tendência central e dispersão. Na figura acima, 5 filhos é um valor discrepante.

Observação:
* O 1o. quartil (inferior) é o valor até ao qual se encontra 25% da amostra ordenada. 
* O 2o. quartil corresponde a mediana, ou seja é o valor até ao qual se encontra 50% da amostra ordenada 
* O 3o. quartil (superior) é o valor até ao qual se encontra 75% da amostra ordenada. 

Veja o calculo da mediana:

````
milsa.Filhos.median()
````

Veja a listagem do número de filhos dos empregados (ordenada) para melhor analisar os resultados e os quartis.

````
milsa.Filhos.sort_values()
````

Note que apenas 20 funcionários possuem número de filhos registrados (NaN é ausência de informação).
A figura abaixo mostra:
* 1o quartil (até o elemento 5 - 25% de 20)
* 2o quartil (até o elemento 10 - 50% de 20)
* 3o quartil (até o elemento 15 - 75% de 20)
