Gráficos são elementos importantes em diversas áreas e também na Matemática.
Um dos usos importantes dos gráficos na matemática é, por exemplo, apoiar a **visualização ou identificação de pontos importantes** em linhas.

Encontrar **raízes de uma função** é um tema bastante trabalhado na nossa formação (em matemática). 

Vale lembrar que, para uma `função f(x)` uma raíz é um valor de "x" para o qual f(x) é igual a zero.

Quando a função é de primeiro ou de segundo grau, existem técnicas relativamente "simples" para encontrar as raízes. Quem não se lembra de `Báskara`?
Mas e para funções mais complexas?

Quais as raízes da função $f(x) = sen(x) + cos(1 + x^2) - 1$?


Uma possibilidade é tentar estimar valores para x e verificar o quanto f(x) se aproxima de zero. E então, traçar um gráfico num intervalo que abranja o valor de x

Note que a figura nos permite verificar que a linha "toca o eixo x", ou seja, cruza o eixo "x" um pouco antes de x=2 e mais ou menos de x=2.5

##Função de primeiro grau
Vamos plotar um gráfico para a função **f(x) = 7 – 3x** para `valores de x` no intervalo [-1;6]. Dizemos que esse intevalo de "x" é domínio da função.

### Gerando valores para x

A função "arange" (`range`: intervalo em inglês) gera uma sequência de valores entre "dois números" (limite inferior e superior)

````
# Gerando uma série de números de -1 a 6
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-1, 6)
x
````

Note que a instrução acima gerou um conjunto de valores entre -1 e 6, mas não incluiu o 6).  

Veja abaixo que `size` retorna o tamanho de x, ou seja, quantos elementos existem em x.

````
print(x.size)
````

### Calculando f(x)
A instrução a seguir calcula o valor de y, ou seja f(x) para todos os 7 valores de x (-1, 0, 1, 2, 3, 4, 5).

````
# calculando o valor de f(x)
y = 7 - 3*x
print(y)       ## mostra y
print(y.size)  ## mostra o número de elementos de y
````

### Exibindo os valores de f(x) associado a cada "x"
Apenas para confirmar, vejamos quanto seria o f(-1):
* y = 7 - (3*-1)
* y = 7 - (-3) 
* y = 10

Assim, f(-1) = 10.

Note que: 
* x[0] = -1   e
* y[0] = 10   ou seja, 
* o primeiro valor de "x" corrsponde ao primeiro valor de "y"

````
print(x[0])   # exibe o primeiro valor de x
print(y[0])   # exibe o primeiro valor de y
````

````
# exibindo na forma f(x) = y
print("f(", x[0], ") = ", y[0])
print("f(", x[1], ") = ", y[1])
````

Podemos exibir todos os valores de f(x) usando a instrução de repetição `enquanto` (`while`). Vejamos:

````
i = 0    # a variável i é usada para percorrer o conjunto x e y
while (i < y.size):   # y.size retorna a quantidade elementos em y
  print("f(", x[i], ") = ", y[i])
  i=i+1
````

É possível também usar a estrutura de repetição "for" para o mesmo problema. Note que a instrução "for" tem passos implícitos, ou seja, esconde algumas coisas, como por exemplo, o incremento "i=i+1".

````
for i in range(y.size):
  print("f(", x[i], ") = ", y[i])
````

Na instrução **"`for i in range(y.size)`"**, a variável `"i"` assume todos os valores entre `0 e y.size` (0 e 7).

````
x = np.arange(-4, 4)             # Gerando uma série de números de -4 a 4
y = x**2 - 3                     # calculando o valor de f(x)
# 0) Montar o Gráfico
plt.plot(x, y,linewidth=2.0, color="blue")     
# 1) Obter as informações gráfico
ax = plt.gca()  # gca significa 'get current axis'
# 2) Retirar a linha da DIREITA (color = none)
ax.spines['right'].set_color('none')
# 3) Retirar a linha de CIMA (color = none)
ax.spines['top'].set_color('none')
# 4) Posicionar a linha de BAIXO, exatamente no "y=0"
ax.spines['bottom'].set_position(('data',0))
# 5) Posicionar a linha da ESQUERDA, exatamente no "x=0"
ax.spines['left'].set_position(('data',0))
# 6) Mostrar o resultado
plt.show()
````

Note que o gráfico fico meio "quadrado". 
Por que será que isso aconteceu?

Veja que foram gerados "poucos" valores de $x$. Com poucos valores de $x$, o gráfico foi "montado" com poucos pontos e por isso ficou meio quadrado. A célula abaixo mostra a quantidade e elementos em x (x.size)

````
x.size
````

Veja abaixo os valores de x.

````
x
````

Vamos gerar mais valores para $x$, usando um incremento menor. No "arange" podemos usar um terceiro parâmetro (opcional) para indicar qual o incremento entre cada elemento. Quando omitimos o parâmetro **o incremento usado é 1**. Por isso, [-4, -3, -2 ...]. Vejamos o que ocorre ao usarmos um **incremento 0.01**.

````
x = np.arange(-4, 4,0.01)
x.size
````

Foram gerados 800 valores  para x, entre -4 e 4, ou seja: [-4, 3.99, 3.98, 3.97, ..., 3.99].
Agora, ao plotarmos novamente o gráfico, veja que ele ficou as curvas suaves.

````
x = np.arange(-4, 4,0.01)     # Gerar os valores de x com intervalo de 0.01
y = x**2 - 3                  # Calcular os valores de f(x) 
# 0) Montar o Gráfico
plt.plot(x, y,linewidth=2.0, color="blue")
# 1) Obter as informações gráfico
ax = plt.gca()  # gca significa 'get current axis'
# 2) Retirar a linha da DIREITA (color = none)
ax.spines['right'].set_color('none')
# 3) Retirar a linha de CIMA (color = none)
ax.spines['top'].set_color('none')
# 4) Posicionar a linha de BAIXO, exatamente no "y=0"
ax.spines['bottom'].set_position(('data',0))
# 5) Posicionar a linha da ESQUERDA, exatamente no "x=0"
ax.spines['left'].set_position(('data',0))
# 6) Mostrar o resultado
plt.show()
````

Pelo gráfico acima as raizes de f(x) estão "próximas" de -2 e 2, mas não tão próximas assim.

### Exibindo f(x) apenas para valores de f(x) "iguais a zero"

````
x = np.arange(-4, 4,0.01)     # Gerar os valores de x com intervalo de 0.01
y = x**2 - 3                  # Calcular os valores de f(x) 
print("Raízes de f(x) no intervalo ", x[0], " e ", x[x.size-1])
for i in range(y.size):
  if (y[i] == 0):
    print("f(", x[i], ") = ", y[i])
````

Note que nenhum valor de f(x) foi exibido acima. Por que será que isso ocorreu?

Vamos então, exibir os valores de f(x) para todos os valores de "x"(800 valores !!!)  e analisar os valores perto de -2.

````
x = np.arange(-4, 4,0.01)     # Gerar os valores de x com intervalo de 0.01
y = x**2 - 3                  # Calcular os valores de f(x) 
for i in range(y.size):  
    print("f(", x[i], ") = ", y[i])
````

Note que os valores de f(x) mais próximos de zero estão ao redor de -1.8, mas de fato, não há nenhum valor de f(x) exatamente igual a 0.0.

Conclui-se que o intervalo de 0.01 não foi suficiente para gerar a "raíz exata", mas esse não é o objetivo aqui. Queremos graficamente enxergar as raízes aproximadas.

### Exibindo f(x) para valores de f(x) perto de zero

De qualquer forma, o trecho abaixo ajuda a listar apenas os valores de x, cujo f(x) esteja entre -0.03 e 0.03.

````
x = np.arange(-4, 4,0.01)     # Gerar os valores de x com intervalo de 0.01
y = x**2 - 3                  # Calcular os valores de f(x) 
for i in range(y.size):  
    if (y[i] >= -0.03) & (y[i] < 0.03):
       print("i=", i, " - f(", x[i], ") = ", y[i])
````

Note pelas informações exibidas acima que as duas raízes estão respectivamente bem próximas a `-1.730` e `1.729`, ou seja, os valoresa de "x" da posição i=227 e i=573.

###Movendo os eixos secundários

Eixos secundários (spines) são as linhas que ligam as marcas de escala do eixo principal, observando os limites da área de dados. Existe quatro deles (superior / inferior / esquerda / direita)

````
#importando a biblioteca numpy
import numpy as np
#importando a biblioteca para gráficos
import matplotlib.pyplot as plt
# Gerando uma série de números de -2 a 4, variando de 0.01
x = np.arange(-2, 4, 0.01)
# calculando o valor de f(x)
# f(x) = sen(x) + cos(1 + x^2) - 1?
y = np.sin(x) + np.cos(1+x**2)-1
plt.plot(x, y,linewidth=2.0, linestyle="-", color="red")
#--------------------------------
ax = plt.gca()  # gca significa 'get current axis'
#retira a linha da DIREITA
ax.spines['right'].set_color('none')
#retira a linha de CIMA
ax.spines['top'].set_color('none')
# posiciona a linha de BAIXO, exatamente no "y=0"
ax.spines['bottom'].set_position(('data',0))
# posiciona a linha da ESQUERDA, exatamente no "x=0"
ax.spines['left'].set_position(('data',0))
````

### Código para identificar as raízes aproximadas

O código abaixo exibe os valores de f(x) entre -0.03 e 0.03

````
x = np.arange(-2, 4, 0.01)
y = np.sin(x) + np.cos(1+x**2)-1
for i in range(y.size):  
    if (y[i] >= -0.03) & (y[i] < 0.03):
       print("i=", i, "- f(", x[i], ") = ", y[i])
````

Os dois valores de x, para o qual é f(x) estão mais próximos de 0 (zero) são os da posição i=394 e i=453.

### Marcando pontos no gráfico

As instruções 5 e 7 abaixo marcam pontos próximos às duas raízes


````
# 1) Gerando uma série de números de -2 a 4, variando de 0.01
x = np.arange(-2, 4, 0.01)
# 2) calculando o valor de f(x)
y = np.sin(x) + np.cos(1+x**2)-1
plt.plot(x, y,linewidth=2.0, linestyle="-", color="red")
# 3) Movendo os eixos secundários
ax = plt.gca()  # gca significa 'get current axis'
#retira a linha da DIREITA
ax.spines['right'].set_color('none')
#retira a linha de CIMA
ax.spines['top'].set_color('none')
# posiciona a linha de BAIXO, exatamente no "y=0"
ax.spines['bottom'].set_position(('data',0))
# posiciona a linha da ESQUERDA, exatamente no "x=0"
ax.spines['left'].set_position(('data',0))
# 4) Verificando raízes aproximadas com base na visualização prévia do gráfico
t = x[394]  # obtem o valor de x da posição 394 (~1.94)
y1= np.sin(t) + np.cos(1+t**2)-1
print("f(",t,")=", y1)
# 5) Marcando em azul o ponto f(t) ~ 0.0
plt.scatter([t, ], [y1, ], 50, color='blue')
# 6) Verificando raízes aproximadas com base na visualização prévia do gráfico
t = x[453]  # obtem o valor de x da posição 453 (~2.53)
y2= np.sin(t) + np.cos(1+t**2)-1
print("f(",t,")=", y2)
# 7) Marcando em verde o ponto f(t) ~ 0.0
plt.scatter([t, ], [y2, ], 50, color='green')
# 8) Mostrando o gráfico
plt.show()
````

Adicionalmente podemos fazer outras anotações. Vide os passos 6 e 9 que geram essas anotações na figura abaixo.

````
# 1) Gerando uma série de números de -2 a 4, variando de 0.01
x = np.arange(-2, 4, 0.01)
# 2) calculando o valor de f(x)
y = np.sin(x) + np.cos(1+x**2)-1
plt.plot(x, y,linewidth=2.0, linestyle="-", color="red")
# 3) Movendo os eixos secundários
ax = plt.gca()  # gca significa 'get current axis'
#retira a linha da DIREITA
ax.spines['right'].set_color('none')
#retira a linha de CIMA
ax.spines['top'].set_color('none')
# posiciona a linha de BAIXO, exatamente no "y=0"
ax.spines['bottom'].set_position(('data',0))
# posiciona a linha da ESQUERDA, exatamente no "x=0"
ax.spines['left'].set_position(('data',0))
# 4) Verificando raízes aproximadas com base na visualização prévia do gráfico
t = x[394]  # obtem o valor de x da posição 394 (~1.94)
y1= np.sin(t) + np.cos(1+t**2)-1
print("f(",t,")=", y1)
# 5) Marcando em azul o ponto f(t) ~ 0.0
plt.scatter([t, ], [y1, ], 50, color='blue')
# 6) Apontando o ponto da primeira raíz
plt.annotate(r'$sin({1.95}) + cos({1+1.95^{2}})$',
            xy=(t, y1), xycoords='data',
            xytext=(-100, +70), textcoords='offset points', fontsize=14,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
# 7) Verificando raízes aproximadas com base na visualização prévia do gráfico
t = x[453]  # obtem o valor de x da posição 453 (~2.53)
y2= np.sin(t) + np.cos(1+t**2)-1
print("f(",t,")=", y2)

# 8) Marcando em vermelho o ponto f(t) ~ 0.0
plt.scatter([t, ], [y2, ], 50, color='green')
# 9) Apontando o ponto da primeira raíz 
plt.annotate(r'$sin({2.54}) + cos({1+2.54^{2}})$',
            xy=(t, y2), xycoords='data',
            xytext=(80, +60), textcoords='offset points', fontsize=14,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
# 8) Mostrando o gráfico
plt.show()
````
